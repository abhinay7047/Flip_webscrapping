```python

# Create an empty DataFrame to store the smartphone data
final_df = pd.DataFrame()

# Loop through the pages from 1 to 44 (45 is not included) on Flipkart search results
for i in range(1, 45):
    # Get the HTML content of the current page
    webpage = requests.get("https://www.flipkart.com/search?q=smartphones&page={}".format(i)).text
    
    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(webpage, "lxml")
    
    # Find all the div elements with the class "_2kHMtA" that contain smartphone information
    all_divs = soup.find_all("div", class_="_2kHMtA")
    
    # Initialize lists to store the data for each smartphone on this page
    image = []
    phonename = []
    price = []
    specification = []
    rating = []
    
    # Loop through each smartphone div and extract relevant information
    for phone in all_divs:
        try:
            # Extract the image URL of the smartphone
            image.append(phone.find("img", class_="_396cs4")["src"])
        except:
            image.append(np.nan)
        
        try:
            # Extract the name of the smartphone
            phonename.append(phone.find("div", class_="_4rR01T").text.strip())
        except:
            phonename.append(np.nan)
        
        try:
            # Extract the price of the smartphone
            price.append(phone.find("div", class_="_30jeq3 _1_WHN1").text.strip())
        except:
            price.append(np.nan)
        
        try:
            # Extract the specifications of the smartphone
            specification.append(phone.find("ul", class_="_1xgFaf").text.strip())
        except:
            specification.append(np.nan)
        
        try:
            # Extract the rating of the smartphone
            rating.append(phone.find("div", class_="_3LWZlK").text.strip())
        except:
            rating.append(np.nan)
    
    # Create a DataFrame for the data extracted from the current page
    df = pd.DataFrame()
    df['Image'] = image
    df['Phone Name'] = phonename
    df['Price'] = price
    df['Specification'] = specification
    df['Rating'] = rating
    
    # Append the data from the current page's DataFrame to the final DataFrame
    final_df = final_df.append(df, ignore_index=True)
    
    # Clear the lists for the next page
    image.clear()
    phonename.clear()
    price.clear()
    specification.clear()
    rating.clear()
```

```python
final_df.to_csv("Flipkart_mobile_data.csv")
```
